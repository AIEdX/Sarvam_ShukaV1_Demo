```python
import transformers
import librosa
from google.colab import files
```

```python
# Load the model pipeline on GPU
pipe = transformers.pipeline(model='sarvamai/shuka_v1', trust_remote_code=True, device=0, torch_dtype='bfloat16')
```

```python
# Upload an audio file
uploaded = files.upload()

# Get the filename of the uploaded audio
audio_file = list(uploaded.keys())[0]
```

```python
# Load the audio file
audio, sr = librosa.load(audio_file, sr=16000)



# Set up the conversation turns
turns = [
    {'role': 'system', 'content': 'Respond naturally and informatively. Repeat the question in a calm professional manner'},
    {'role': 'user', 'content': '<|audio|>'}
]

```

```python
# Run the pipeline
result = pipe({'audio': audio, 'turns': turns, 'sampling_rate': sr}, max_new_tokens=512)

# Print the result
print(result)
```
